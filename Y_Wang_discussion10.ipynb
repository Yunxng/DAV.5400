```python
# load the pandas + numpy libraries
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
from pandas import Series,DataFrame
# load the Series and DataFrame functions 
```

```python
# read contents of a csv file that we build from GitHub into a Pandas DataFrame
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/'
                 'bridges/bridges.data.version1', header=None)
#Set columns name
df.columns=['IDENTIF','RIVER','LOCATION','ERECTED','PURPOSE','LENGTH','LANES','CLEAR-G','T-OR-D','MATERIAL','SPAN','REL-L','TYPE']
```

Task1 

```python
# use the size() function to tell us how many data values belong to each grouping
df1=pd.DataFrame(df.groupby(['RIVER','PURPOSE', 'MATERIAL']).size())
# give the '0' column a meaningful label
df1.rename(columns={0:'How Many?'}, inplace = True)
df1
```

Task2

```python
#Set a'value fo '?' in "LENGTH" as NaN and change the value as numeric
df['LENGTH'][df['LENGTH']=='?'] = np.nan
df['LENGTH'] = df['LENGTH'].astype(float)
#applying the mean() function to the group
df2=pd.DataFrame(df.groupby([df['PURPOSE'],df['MATERIAL']])['LENGTH'].mean())
# give the 'LENGTH' column a meaningful label
df2.rename(columns={'LENGTH':'Average Length'}, inplace = True)
df2
```

```python
quartiles = pd.cut(df.ERECTED, 4)
# define a function that specifies the aggregating functions we want to apply
# to the length values contained within each of our data1 quartile cuts
def get_stats(group):
    return {'Average Length': group.mean(), 'Count': group.count(),
            'Max Length': group.max(), 'Min Length': group.min()}
# group the length column's values relative to the erected quartile cuts we made
grouped = df.LENGTH.groupby(quartiles)
# apply the get_stats function to the grouped length values
grouped.apply(get_stats).unstack()
```
